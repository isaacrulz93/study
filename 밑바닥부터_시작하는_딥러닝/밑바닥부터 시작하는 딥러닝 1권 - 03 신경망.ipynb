### 3.1.2 퍼셉트론 복습

### 기본적인 퍼셉트론의 개형도

퍼셉트론은 기본적으로 위의 꼴을 한다고 설명했던지만, 하지만 우리는 $\theta$ 라는 친구를 편향이라 두고 

<center>
$ y=
\begin{cases}
0 & (b + w_1 x_1 + w_2 x_2) \le 0 \\
1 & (b + w_1 x_1 + w_2 x_2) > 0 \\
\end{cases}
$
</center>
꼴로 정리를 했었잖아? 그리고 다시 한번 말하면 편향 b는 뉴런이 <b>어느 수준에서 활성화 할 값인지?</b>를 결정하는 거라 했고,<br>
가중치 w는 각 요인을 <b>얼마나 중요하게 생각할건지?</b> 라고 얘기했었어.<br>
무튼, 위의 그림으론 우리의 편향이 표현이 안된단 말이지.<br>
그래서 편향을 추가한 그림을 그리면 다음과 같아

이제 위의 식을 더 간단하게 써보자.
<center>
$y = h(b+w_1 x_1 + w_2 x_2)$
</center>

<center>
$h(x) = 
\begin{cases}
0 & (x \le 0) \\
1 & (x > 0) \\
\end{cases}$
</center>

로 바꿔서 표현 할 수 있어<br>
이러면 입력신호의 총 합$(b+w_1 x_1 + w_2 x_2)$이 $h(x)$의 $x$로 작용되고, $h(x)$함수의 출력을 통해 y값을 결정하게 됐잖아?<br>
여태까지 $y=$ &#126;&#126;&#126; 꼴로 잘 써오다 갑자기 h(x)가 왜 등장하냐? 앞으로 설명할 활성화 함수라는 개념을 꺼내기 위해 그래.

### 3.1.3 활성화 함수 - Activation Function

위의 $h(x)$처럼 입력신호의 총 합을 출력함수로 변환시켜 주는 함수를 앞으로 <b>활성화 함수</b> 라고 할거야.<br>
이름이 활성화 함수인거처럼, 입력받은걸로 얘를 활성화 시킬까 말까? 를 결정해주는 함수야<br><br>

<center>
$h(x) = 
\begin{cases}
0 & (x \le 0) \\
1 & (x > 0) \\
\end{cases}$
</center><br><br>

<center>설명한 부분이 조금 햇갈릴 수 있으니까, 조금 바꿔 쓰면</center><br><br>

<center>
   $a = b +w_1 x_1 + w_2 x_2$
    <br>
   $y = h(a)$
</center>
<br>

로 바꿔 쓸 수 있어. 이걸 그림으로 표현하면 다음과 같은 그림이 나와. 좀 더 이해하기 쉬워졌으려나.<br>저 노드 안에서 무슨 일이 일어났는지 보여주잖아

### 3.2.1 Sigmoid 함수

Sigmoid 함수는 신경망에서 굉장히 많이 이용되는 함수 중 하나야<br><br>
<center>$h(x) = {{1}\over{1+exp(-x)}}$</center><br>
꼴로 쓰여. $exp(-x)$는 자연상수 $e$를 밑으로 하고 지수가 $-x$, 즉 $e^{-x}$를 의미해

시그모이드는 신경망에서 많이 쓰이는 함수고, 왜 많이 쓰이는지는 차차 설명해줄게

### 3.2.2 계단함수 구현하기

전에 우리가 퍼셉트론을 눈으로 비교했었었잖아, 이번에도 계단함수와 시그모이드를 직접 짜고, 눈으로 비교하면서 차이를 알아보자

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
print(step_function(5.1))
print(step_function(-1.1))
print(step_function(0))
```

하지만 지금 우리가 만든 계단함수는 입력값을 하나밖에 못받는단 말이지? 앞으로 우리는 대부분의 데이터 처리를 행렬, 벡터 꼴로 처리를 할텐데?
그러면 이 함수가 배열을 입력으로 받을수 있게도 만들어보자

```python
import numpy as np

def step_function_for_numpy(x):
    y = x > 0
    return y.astype(np.int)
```

```python
x = np.array([1,-1,2,-3,4])
step_function_for_numpy(x)
```

### 3.2.4 시그모이드 함수

```python
def sigmoid(x):
    return 1/ (1+np.exp(-x))
```

끝이야. 쉽지? 그리고 np.를 이용한 수식이 들어갔기때문에 입력값을 바로 넘파이 배열 형태로 넣어줘도 확인할수있어.<br>
이건 넘파이의 브로드캐스팅기능 떄문이야

```python
x = np.array([-1,1,2,-4])
sigmoid(x)
```

```python
x = np.arange(-10,10,0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.show()
```

TMI로 Sigmoid 라는 뜻은, Sigma의 모양을 닮았다. 해서 Sigmoid야. $\Sigma$ 말고 그냥 영어 알파벳 S에 대응해줘. 그래서 Sigmoid는 S자 모양으로 나온다는걸 기억해주면 좋아

### 3.2.5 Sigmoid 와 계단함수의 비교

```python
x = np.arange(-5,5,0.1)
y1 = step_function_for_numpy(x)
y2 = sigmoid(x)

plt.plot(x,y1,'--',c='b',label='step_function')
plt.plot(x,y2,c='r',label='sigmoid_function')
plt.legend()
plt.show()
```

보면 Sigmoid와 계단함수는 굉장히 유사하게 생겼어.<br>
둘의 공통점을 먼저 볼까? 둘다 0~1사이에서 값이 출력되고, 입력값이 크면 큰값을 출력해<br>
하지만 Sigmoid 함수는 중간이 '매끄럽게' 이어져있는 반면,<br>
계단함수는 딱딱하게 끊겨있지. 여기서 '부드럽게' 이어졌다는건 굉장히 중요한 부분이야.<br>
이 '부드럽게' 이어져있기 때문에 sigmoid 함수는 모든점에서 '미분'이 가능해지거든.<br>
갑자기 미분이 왜 나오냐고? 그건 차차 설명해줄게

추가적으로, Sigmoid는<br>
- 0~1사이의 값을 갖기떄문에, 하나의 '확률'로 표현 가능하고,<br>
- 궁극적으로는 0쪽이냐, 1쪽이냐로 나뉘기 떄문에 이진분류(Binary Classification)의 마지막 출력층 활성화 함수로 주로 쓰여

### 3.2.6 비선형함수

또 둘의 공통점은 비선형함수라는거야. 직선으로 표현이 불가능해.<br>
그리고 앞으로 배울 신경망은 활성화 함수로 비선형함수를 이용해야해.<br>
왜냐고? 우리가 지금 할게 '신경망'이고 중간에 '층'이 존재한다고 했잖아? <br>
근데 활성화 함수로 선형함수를 쓰면 여태까지 쌓아올린 층이 의미가 없어.<br>
선형적인 함수는 겹겹이 쌓아도 앞의 계수만 변하고, 선형이란 구조 자체는 변함없기 떄문이야.<br><br>

예를들여, $h(x) = ax$ 가 있다고 하자. $h(h(h(x)))$ 라고 겹겹이 쌓아도, 결국엔 $a^3 x$라는 하나의 직선으로 표현이 가능하게 되잖아.<br>
그럼 애초에 $h(x) = a^3 x$ 로 해버렸으면 되는거거든. 그러면 신경망을 쓰는 의미가 없어져<br>
그래서 신경망은 활성화함수로 비선형함수를 써야해

### 3.2.7 ReLU 함수

ReLU는 Rectified Linear Unit의 약자로, 정류된 선형 유닛? 정도로 알고 있으면 될거같아.<br>
중요한건, 입력이 0을 넘으면 입력을 그대로 출력하고, 0보다 작으면 0을 출력하는 함수라는거야<br>
<center>
$h(x) = 
\begin{cases}
x & (x>0) \\
0 & (x\le0)\\
\end{cases}
$
</center>
로 수식으로 표현할 수 있어. 그려보면 다음과 같아

```python
def ReLU(x):
    return np.maximum(0,x)
```

np.maximum은 둘중 큰 값을 출력해주는 넘파이의 함수야<br>
0이나 x 중 큰값을 뽑아주세요! 라는 뜻이지

```python
print(ReLU(0))
print(ReLU(5))
print(ReLU(-1))
x = np.array([1,2,3,-1,0,-2])
print(ReLU(x))
```

```python
x = np.arange(-5,5,0.1)
y = ReLU(x)

plt.plot(x,y,label = 'ReLU')
plt.figure(figsize=(10,12))
plt.show()
```

### 3.2.8 tanh - 하이퍼볼릭탄젠트 함수(Hyperbolic tangent function)

활성화 함수로 하이퍼볼릭 탄젠트 함수를 이용하는거야. 수식은<br>
<h3><center>$
y ={{e^x - e^{-x}}\over{e^x + e^{-x}}}
$</center></h3><br>
를 이용해

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
print(tanh(3))
print(tanh(-3))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
x = np.arange(-5,5,0.1)
y1 = tanh(x)
y2 = sigmoid(x)

plt.plot(x,y1,c = 'r')
plt.plot(x,y2,c = 'b')
plt.show
```

```python
x = np.arange(-5,5,0.1)
y = tanh(x)

plt.plot(x,y)
plt.show()
```

이 친구는 시그모이드와 유사한 모양을 띄어

### 3.2.9 Leaky ReLU

위에서 본 ReLU는 0이하의 값은 싹 다 0으로 바꿔버렸었잖아?<br>
Leaky ReLU는 0보다 작은 값도 '살려는 드릴게'의 느낌으로, 작은 값으로 변환시켜줘
수식으로 쓰면 이렇게 쓸 수 있고,<br>0보다 작을때의 기울기는 임의로 변경해도 돼. 도식화시키면 다음과 같아
<center>$
f(x) = 
\begin{cases}
0.01x & x<0 \\
x & x \ge 0
\end{cases}
$
</center>

```python
def LeakyReLU(x):
    a = 0.01
    return np.maximum(a*x,x)
# 음수가 되면 1보다 작은 값을 곱해준게 더 커지는 논리를 이용한거야
```

```python
print(LeakyReLU(1))
print(LeakyReLU(2))
print(LeakyReLU(-3))
```

```python
x = np.arange(-10,2,0.1)
y = LeakyReLU(x)

plt.plot(x,y)
plt.show()
```

### 3.2.10 ELU (Expotential Linear Units)

ELU 또한 ReLU 함수의 단점을 보완해주기 위해 만들어진 함수로, 수식으로 쓰면 다음과 같아.<br>
이 함수는 $\alpha$가 1일때 0에서 매끄럽다정도만 알아두자
<center>$
f(\alpha,x) = 
\begin{cases}
\alpha (e^x -1) & (x \le 0) \\
x & (x > 0) \\
\end{cases}$
<\center>

```python
def ELU(x):
    a = 1
    return (x>0)*x + (x<=0) *a*(np.exp(x)-1)
```

```python
print(ELU(1))
print(ELU(3))
print(ELU(-2))
x = np.array([-1,2,-3,4])
print(ELU(x))
```

```python
x = np.arange(-5,5,0.1)
y = ELU(x)

plt.plot(x,y)
plt.show()
```

- ReLU
- Sigmoid<br>

### 3.3.1 다차원 배열

```python
import numpy as np

A= np.array([1,2,3,4])
print(A)

print('A의 차원 :',np.ndim(A))

print('A의 모양(x by x인지?) :',A.shape)
```

지금 A는 한줄짜린데 A.shape에서 4 하나만 출력해줘도 될거같은데 (4,)로 굳이 튜플로 출력해줬잖아?<br>

### 3.3.2 행렬의 곱

2X2 행렬에서 행렬의 곱셈은 다음과 같이 진행돼. <b>A</b>와 <b>B</b>라는 행렬을 곱해서 새로운 행렬 <b>C</b>를 만들어내는거야.<br>
참고로 표기할때 단일 원소는 a 이런식으로 표현한다면, 벡터는 <b>a</b>, 행렬은 <b>A</b> 꼴로 자주표기하니까 알아두면 좋아<br>
행렬 곱셈은 넘파이를 통해 굉장히 쉽게 구할 수 있어

```python
import numpy as np

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print('A의 형태 :',A.shape)
print('B의 형태 :',B.shape)

print('A와 B의 곱 :\n',np.dot(A,B))
print('B와 A의 곱 :\n',np.dot(B,A))
print('')
print('')
print('A*B를 바로 하면? : Hadamard Product(아다마르 곱)이 나온다\n',A*B)
```

행렬은 곱하는 순서에 따라 값이 변할 수 있다는건 알고있어야해. 행렬은 연산 순서에 따라 결과가 달라. 경우에 따라선 모양도 바뀌어<br><br>

덧셈에 대한 역원 : <b>-A</b>, A의 모든 원소에 -1을 곱해준거야<br>
곱셈에 대한 항등원 : <b>E</b> (단위행렬), A가 $n$ x $n$ 행렬일때, 대각선의 값은 다 1이고 나머지는 0인 행렬이야<br>

### 3.4.3 구현 정리

이번엔 네트워크 전체를 그냥 하나의 함수로 만들어버려서, 입력을 받으면 바로 그 값들을 출력할 수 있게 해볼꺼야

이건 네트워크의 구현이야. 비어있는 네트워크 딕셔너리를 만들어줘서, 정의할 키에 값(Value)들을 넣어주는거야

```python
def network():

    network = {}
    #첫번째 레이어
    network['W1'] = np.array([[0.1,0.3,0.5],
                              [0.2,0.4,0.6],
                              [0.3,0.5,0.7]])
    
    network['B1'] = np.array([1,2,3])

    #두번쨰 레이어
    network['W2'] = np.array([[0.2,0.4,0.1],
                              [0.1,0.6,0.2],
                              [0.5,0.8,0.1]])
    
    network['B2'] = np.array([1,1,1])
  
    #세번째 레이어
    network['W3'] = np.array([[0.1,0.3],
                              [-0.1,-0.5],
                              [0.3,0.2]])
    
    network['B3'] = np.array([1,1])
 
    #네번째 레이어
    network['W4'] = np.array([[0.1,0.2],
                              [0.3,0.5]])
    network['B4'] = np.array([1,1])

    return network
```

이건 계산 과정의 구현이야. 이름이 forward인 이유는 차례대로 계산하는 순전파 과정이기 때문에 그래.<br>
차례대로 계산하는게 당연한거 아니야? 라고 생각 할 수있겠지만, 조만간 딥러닝의 꽃인 역전파를 배우게 될텐데, 그와 구별하기 위해 순전파라고 해

```python
def forward(network,x):
    W1,W2,W3,W4 = network['W1'],network['W2'],network['W3'],network['W4']
    B1,B2,B3,B4 = network['B1'],network['B2'],network['B3'],network['B4']

    A1 = np.dot(x,W1) + B1
    Z1 = sigmoid(A1)

    A2 = np.dot(Z1,W2) + B2
    Z2 = sigmoid(A2)

    A3 = np.dot(Z2,W3) + B3
    Z3 = sigmoid(A3)
  
    A4 = np.dot(Z3,W4) + B4
    y = identity_function(A4)
  
    return y
```

```python
network = network()
x = np.array([0.3,1.3,-2.2])
y = forward(network,x)
print(y)
```

### 3.5.1 항등함수와 소프트맥스(Softmax Function) 함수 구현

항등함수는 앞에 나왔던 identity_function(x)야. 이제 소프트맥스 함수에 대해 알아보자.<br><br>

소프트맥스 함수는 <b>여러 클래스로 분류되어있는걸 예측</b>하기 위해 주로 사용 되는 함수야 (Multi Class Classification)<br><br>

입력값의 영향을 크게 받아서 입력값이 커지면 그만큼 출력값도 커져.<br>
게다가 출력된 결과물의 전체 합이 1이라 출력값 자체를 하나의 **확률**로 사용 가능해<br>
수식은 다음과 같아

<h1><center>
$y_k = {{{exp(a_k)}\over\sum_{i=1}^n exp(a_i)}}$
</center></h1>

여기서 $n$은 출력층의 뉴련수<br>
$y_k$는 $k$번째 출력<br>
$a_k$는 입력신호 $a_k$를 뜻해.


아까 softmax 함수의 총 합은 1이라고 했는데,<br><br>
<font size = '5em'> ${{exp(a_1)}\over\sum_{i=1}^n exp(a_i)} +...+{{exp(a_n)}\over\sum_{i=1}^n exp(a_i)} $ </font> 을 하면
<font size = '5em'>${\sum_{i=1}^n exp(a_i)}\over{\sum_{i=1}^n exp(a_i)}$</font>이라 총합이 1이 나오는거야

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([1.5,1.2,4.5])
print('a의 값 :',a)
print('exp(a)의 값 :',np.exp(a))
print('softmax(a)의 값 :',softmax(a))
print('softmax(a)의 합 :',np.sum(softmax(a)))
```

보면 알겠지만 입력값은 [1.5,1.2,4.5] 였음에도 불구하고, softmax를 통해 나온값은 [0.04,0.03,0.92] 로 굉장히 큰 차이를 보여<br>
특히 아까 처음에 softmax의 값은 확률로도 사용 가능하다는 점을 생각해보면<br>
입력값은 3배? 정도밖에 차이가 나지 않지만, softmax를 통해 나온 값은 20~30배에 육박하는 차이를 보이니까.<br>
<b>정답일꺼같은놈을 더 정답인거처럼</b> 만들어준다고 생각하면 될거같아<br>
그리고 정답인놈을 더 정답인거처럼 만들어준다 했지만,<br>
softmax 내부에 쓰인 exp 함수가 단조증가함수(monotone-increasing)이기 때문에 각 수의 대소관계엔 변함이 없다는점도 주의해주면 좋아

* 단조증가는 $a \le b$ 일때, $f(a) \le f(b)$을 만족시켜줘

### 3.5.2 소프트맥스 함수의 문제점

softmax 함수 자체는 문제없이 깔끔한 연산방법인게 맞는데, 우리가 이걸 지금 파이썬을 통해 컴퓨터로 계산하는거잖아?<br>
그러면서 문제가 하나 생겨. 오버플로(overflow)라는 문제야. 컴퓨터가 계산할 수 있는 양을 넘어서버리는거지.<br>
지수함수를 이용하기 때문에 생기는 문제야. 예를 들어서 보면 $e^{10}$은 20,000이 넘고, $e^{100}$은 0이 40개가 넘어가.<br>
컴퓨터가 감당 가능한 연산을 넘어버릴수 있다는거지. 이걸 해결하기 위해선 $exp$ 함수의 성질을 잘 이해해서 약간의 조작을 가해주면 돼 

<font size = '4em'>$y_k = {{{exp(a_k)}\over\sum_{i=1}^n exp(a_i)}}$</font> 일때, 분자/분모에 같은 숫자를 곱해주면 그 값은 변하지 않지? 그 점을 이용해
<br><br>
<font size = '4em'>${{C{exp(a_k)}\over C \sum_{i=1}^n exp(a_i)}}$</font> 꼴로 변환시켜줘. 그 다음엔
<br><br>
<font size = '4em'>$Ce^x = e^{x+\log C}$</font>라는 점을 이용해서 <font size = '4em'>${{{exp(a_k + \log C)}\over \sum_{i=1}^n exp(a_i + \log C)}}$</font> 꼴로 바꿔 쓸 수 있단말이지
<br><br>
근데 사실 $\log C$라는건 내가 어떤 숫자 C를 설정하냐에따라 값이 변하는거기때문에, $C'$으로 표기가 가능해. 그러면 softmax의 변형은 다음과 같아
<br><br>
<font size = '4em'>
<center>$y_k = {{exp(a_k + C')} \over {\sum_{i=1} ^n exp(a_i + C')}}$</center>
</font>
<br><br>
이게 무슨뜻이냐면, softmax를 계산할때 어떤 숫자를 더하거나 빼도 결과는 변하지 않는다는거야.<br>
하지만 우리가 이 가공을 한 이유는 오버플로를 해결하기 위한거였으니까, 입력신호값중에 제일 큰 값을 빼주는게 가장 일반적이야.<br>

```python
a = np.array([1000,1001,1002])
softmax(a)
```

```python
c = 1000
softmax(a - c)
```

```python
c = np.max(a)
softmax(a-c)
```

위에선 "RuntimeWarning: overflow encountered in exp" 이런 오버플로 문제가 발생했었지만, 적절한 조치를 취해 오버플로를 해결했어.<br>
이를 기반으로 새로운 softmax 함수를 써보면 오버플로문제를 해결한걸 볼 수 있어

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([1000,1001,1002])
softmax(a)
```

### 3.6.2 신경망의 추론 처리

정리가 잘 안되지? 어려울수 있지만, 일단 여기까지 왔으니까 이 악물고 끝까지 가보자

```python
#다시 불러오기 ( 뭐 쓰는지 보라고)
import tensorflow as tf
import numpy as np
import pickle

def sigmoid(x):
    return 1 / (1+np.exp(-x))

def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y


def flatten_for_mnist(x):
    temp = np.zeros((x.shape[0],x[0].size))
    
    for idx,data in enumerate(x):
        temp[idx][:] = data.flatten()
        #temp[idx,:] = data.flatten()
        
    return temp
```

```python
#데이터 불러오기
mnist = tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data()
x_train, x_test =flatten_for_mnist(x_train)/255.0, flatten_for_mnist(x_test)/255.0   #255로 나누는건 정규화를 위해서야
#y_test = flatten_for_mnist(y_test)

#네트워크 설정 (이미 저장된 가중치값을 이용할거야)
def init_network():
    with open('sample_weight.pkl','rb') as f:
        network = pickle.load(f)
    return network

def predict(network,x):
    W1,W2,W3 = network['W1'],network['W2'],network['W3']
    b1,b2,b3 = network['b1'],network['b2'],network['b3']
    
    a1 = np.dot(x,W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1,W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2,W3) + b3
    t = softmax(a3)
    
    return t
```

```python
#정답률을 확인하는 과정

x,y = x_test,y_test
network = init_network()

accuracy_cnt = 0
for i in range(len(x)):
    t = predict(network,x[i])
    p = np.argmax(t)
    if p== y[i]:
        accuracy_cnt +=1
        
print('Accuracy :' + str(float(accuracy_cnt)/len(x)))
```

저장된 가중치값을 이용하긴 했지만, 위에서 구현한 3층 신경망을 기반으로 mnist데이터를 확인해 볼 수 있었어. 정확도는 0.9352가 나왔네<br>

아 지금 네트워크를 거치며 이런 형태로 계산이 됐구나 를 확인 할 수 있어.<br>
28* 28을 펼친 784개를 입력받아 쭉 과정을 거치고, 마지막에 숫자 10개로 분류하기 위해 Y가 10인것도 잊지 말고<br>

근대 이건 지금 x의 행 하나를 집어넣었을때 생기는 일이잖아? 만약에 x의 행을 100개를 predict 함수안에 집어넣으면?

X의 형상이 100 x 784 로 바뀌면서, Y에는 100개의 출력이 한번에 나오게 돼<br>
이렇게 데이터를 묶여있는 묶음을 배치(batch)라고 해. 영어로 일괄처리됐다 라는 뜻이야.<br>
왜 이렇게 하냐고? 컴퓨터가 계산하기 더 좋대

```python
x,y = x_test,y_test
network = init_network()

batch_size = 100
accuracy_cnt = 0

for i in range(0,len(x),batch_size):
    x_batch = x[i:i+batch_size]
    y_batch = predict(network,x_batch)
    p = np.argmax(y_batch, axis=1)
    accuracy_cnt += np.sum((p == y[i:i+batch_size]))
    
print('Accuracy :' + str( float(accuracy_cnt) / len(x)))
```

휴 일단 끝!!!!<br>
무슨일인지 한번 보자<br>

for 문 안에 보면,
범위가 0~ x의 개수만큼 배치사이즈 단위로 읽어달라!<br>
그 다음 배치는 읽기 시작한 순서부터 배치크기만큼 읽어와라!<br>
y배치는 x배치를 예측한 값이다!<br>
p는 y배치에서 가장 큰 값을 행 별로 뽑아줘! 를 말하는거고,(확률로 나오고, 뽑혀나오는건 인덱스가 나와)<br>
p == y[i:i+batch_size] 는 == 연산자니까 참 거짓으로 나오겠지? 이건 0과 1이고<br>
np.sum을 통해 그걸 다 더해줘서 accuracy_cnt에다 담아줘. 참고로 지금 한건 배치 하나에 대한 것이기 때문에, 배치를 반복할때마다 쌓일거야
