### 0.2 데이터의 종류

앞으로 데이터에 대해 이야기를 많이 할텐데, 먼저 데이터가 뭔지부터 알아보자

### 0.2.1 형식에 따른 분류
먼저 데이터의 형식에 따른 분류를 보면, 정형데이터와 비정형데이터로 나뉘게 돼

* 정형데이터(Structured Data)<br>
정형 데이터는 <b>정해진 규칙</b>에 따라 있는 데이터를 뜻해. 엄밀히말하면 아니지만, 그냥 엑셀 차트에 넣을수있는가? 를 생각해보면 될거같아 <br><br>

* 비정형데이터(Unstructured Data)<br>
비정형데이터는, 말 그대로 정형데이터가 아니란 뜻이야.<br>
뭐.. 텍스트, 사진, 영상, 소리 등등 대부분의 데이터들이 여기 속해<br>


### 0.2.2 척도에 따른 분류
척도(Scale)에 따라서 나누면, 질적데이터와 양적데이터로 나눌 수 있어<br>
 (척도란 일정한 규칙에 따라 측정대상에 적용하도록 만들어진 기호나 숫자야)

* 질적 데이터(Qualitative Data) - 질적데이터는 특징같은것을 나타내주는 데이터야. 명목척도와 순서척도로 나눌 수 있어
 * 명목척도(Nominal Data)<br>
 명목척도는 남자, 여자와 같은 분류를 할때 남자=1, 여자=2로 분류하는것 같은 방법이야. 여기서 숫자는 대상을 상징하지, 수로써의 의미는 없어<br><br>
 * 순서척도(Ordinal Data)<br>
 순서척도는 순서를 나타내는 척도야. '올림픽 100m 달리기 순위' 처럼 순서를 수치화 한거지. 이런 데이터는 숫자로의 특징도 지니고 있긴 하지만,<br> 연산을 할 수 없어. 1등이랑 3등을 더한다고 4등이 될 수 없잖아<br><br>

 ### 0.2.2.1 속성에 따른 분류
* 양적 데이터(Quantitive Data) - 양적데이터는 쉽게말하면, 얼마나? 라고 물었을때 대답해 줄 수 있는 데이터야. 연속형과 범주형으로 나눌 수 있어
 * 연속형데이터(Continuous Data) - 키,몸무게, 시간 처럼 연속적인 수치로 표현된 자료들을 뜻해. 연속적이기 때문에 아나로그라고 볼 수 있어<br><br>
 * 범주형데이터(Categorical Data) - 나이, 인원수 처럼 명확하게 구별할 수 있는 데이터의 형태야

강화학습은 시뮬레이션을 통해 지금 내가 한 선택이 나중에 최적의 선택이 되도록 학습하는거야.<br>
