## 2.2 논리게이트

대표적인 논리게이트
- AND

- NAND

우리는 지금 여기서 AND,OR,NAND에 대해서만 알아볼거야. NOT은 부정해준다는거고, NOR은 NOT OR이야

## 2.2.1 AND 게이트

AND 게이트는 불 대수(BOOLEAN ALGEBRA : 참 거짓을 1과 0으로 대입)에서, 모든 입력이 참(컴퓨터는 1)일때 참(1)의 연산을 출력하는 연산자야

###  AND 게이트의 진리표

이를 퍼셉트론으로 표현하기 위해 적절한 $w_1, w_2, \theta$ 값을 찾아보자<br>
어떤 함수에 $(0,0),(1,0),(0,1),(1,1)$ 을 넣었을때 출력이 $(0,0,0,1)$이 나오도록 조절해야해<br>
어떤 함수는, 위에서 보여준 함수의 형태를 따르기로 하고, 여기에 맞는 숫자를 찾아보자<br>
 예를들면 위의 꼴엔 $[w_1, w_2, \theta] = [1,1,1]$을 넣거나,
 $[w_1, w_2, \theta] = [0.3,0.7,0.8] $ 넣어주면 AND 연산자를 만족시켜줘

## 2.2.2 NAND 게이트와 OR 게이트

### NAND 게이트

 NAND는 NOT AND 를 줄인말로, AND 논리의 부정. AND와 반대의 출력을 내주는 게이트야.<br>
 입력이 모두 1일때만 0을 출력하고 나머지 경우엔 모두 1을 출력해줘서<br>
 AND와는 완전 반대되는 결과를 얻을 수 있어

### NAND 게이트의 진리표

AND에서 해봤으니까, 이번에도 마찬가지로 위의 식에 적절한 값을 찾아줘보자<br>
예를 들면 $(w_1, w_2, \theta) = (-0.5,-0.5,-0.7)$ 를 집어넣으면 NAND게이트가 만들어져.<br>
사실 AND게이트를 만들때 사용했던 매개변수에 부호만 바꿔줘도 만들 수 있어

### OR 게이트

OR는 입력신호중 하나라도 1이 있으면(모두 0이 아니면) 출력이 1이 되는 논리게이트야<br>

### OR게이트의 진리표

여태까지 퍼셉트론으로 기본적인 논리 연산자 AND, NAND, OR를 만들어봤어!<br>
3가지 게이트 모두 기본적인 구조는 똑같이 생겼지만,<br>
우리가 매개변수($x_1,x_2,\theta$)를 어떻게 조절하냐에 따라 값이 변한다구

혹시 헷갈릴까봐 쓰는데, 우리는 지금 '퍼셉트론' 이라는 알고리즘을 통해 '논리게이트'를 구현하고 있는거야.<br>
퍼셉트론은 논리게이트랑 관련없어

## 2.3 파이썬으로 퍼셉트론을 활용한 논리게이트 구현하기

#### AND 게이트

```python
def AND(x1,x2):
    # AND이므로, 입력이 (1,1)일때만 1이 출력되도록 숫자를 잘 조절해보자
    w1,w2,theta = 0.6, 0.2, 0.7 
    if w1*x1 + w2*x2 <= theta:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

#### NAND 게이트

```python
def NAND(x1,x2):
    w1,w2,theta = 0.5, 0.5, 0.7
    if w1*x1 + w2*x2 >= theta:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

#### OR게이트

```python
def OR(x1,x2):
    w1,w2,theta = 0.9,0.7,0.6
    if w1*x1 + w2*x2 <= theta :
        return 0
    else:
        return 1
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

## 2.3.2 가중치와 편향

위에서 아래의 꼴로 퍼셉트론을 표현해 봤잖아? <br><br>
<center>
$ y=
\begin{cases}
0 & (w_1 x_1 + w_2 x_2) \le \theta \\
1 & (w_1 x_1 + w_2 x_2) > \theta \\
\end{cases}
$
</center>
<br><br>

여기서 $\theta$를 $-b$ 로 치환하면, 위의 공식을 <br><br>
<center>
$ y=
\begin{cases}
0 & (b + w_1 x_1 + w_2 x_2) \le 0 \\
1 & (b + w_1 x_1 + w_2 x_2) > 0 \\
\end{cases}
$
</center>
<br><br>

꼴로 바꿀 수 있게 되면서, 이제부턴 $b$를 <b>편향</b>이라고 할거야. $w_1,w_2$는 그대로 가중치라 할꺼고.<br>
이러면 이제 퍼셉트론은<br><br>
<center>'입력신호에 가중치를 곱하고, 편향을 더해줬을떄 0을 기준으로 넘으면 1,아니면 0을 뱉는 분류기'</center><br><br>
라고 바꿔서 말할 수 있게 돼<br>
이번엔 좀 다른 방법으로 위의 게이트들을 구현해보자

### AND의 새로운 작성법

```python
def AND2(x1,x2):
    input = np.array([x1,x2]) #들어온 입력들을 넘파이 배열로 만들어준다
    weight = np.array([0.4,0.6]) #넘파이로 계산하기 위해 가중치도 넘파이 배열로 생성
    bias = -0.7 #편향은 한번 더해줄거라 그냥 일반 숫자면 돼
    value = np.dot(weight,input) + bias
    if value > 0:
        return 1
    else:
        return 0
```

```python
print(AND2(0,0))
print(AND2(1,0))
print(AND2(0,1))
print(AND2(1,1))
print(AND2(2,0))
print(AND2(0.5,0.5))
```

### AND 눈으로 보기

```python
x1 = np.arange(-2,2,0.01)
bias = -0.7

#위의 식은 w1x1 + w2x2 + b <0 꼴로 써있었지만, 그래프로 나타내기위해 x2에 대한 식으로 정리한 과정이야

# x2 =  (-w1x1 -b)/w2
x2 = (-0.4 * x1 - bias) / 0.6
plt.axvline(x=0,color = 'black')
plt.axhline(y=0,color = 'black')

plt.plot(x1,x2,'--')
plt.scatter(0,0,color='g',marker='x',s=150)
plt.scatter(1,0,color='g',marker='x',s=150)
plt.scatter(0,1,color='g',marker='x',s=150)
plt.scatter(1,1,color='red',marker='o',s=150)
plt.xlim(-0.5,2)
plt.ylim(-0.5,2)
plt.show()
```

여기서 알아줘야할점은,<br> <b> 가중치</b> : 각각의 입력들에 '넌 얼마나 중요한애니?' 의 정도를 조절하는 매개변수고, <br>

<b>편향</b> : '이 뉴런을 어느정도로 쉽게 활성화 시킬건데?'를 조절하는 매개변수야<br><br>
마저 위의 방법으로 NAND와 OR도 구현해보자

### NAND와 OR의 새로운 작성법

```python
def NAND2(x1,x2):
    input = np.array([x1,x2])
    weight = np.array([-0.4,-0.3]) 
    bias = 0.6 
    value = np.sum(weight*input) + bias
    if value > 0:
        return 1
    else:
        return 0
```

```python
print(NAND2(0,0))
print(NAND2(1,0))
print(NAND2(0,1))
print(NAND2(1,1))
```

### NAND 눈으로 보기

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = 0.6

#위의 식은 w1x1 + w2x2 + b <0 임을 묻는 질문이였을때, x2에 대한 식으로 정리한 과정
x2 = (0.4 * x1 - bias) / -0.3
plt.axvline(x=0,color = 'black')
plt.axhline(y=0,color = 'black')

plt.plot(x1,x2,'--')
plt.scatter(0,0,color='red',marker='o',s=150)
plt.scatter(1,0,color='red',marker='o',s=150)
plt.scatter(0,1,color='red',marker='o',s=150)
plt.scatter(1,1,color='green',marker='x',s=150)
plt.xlim(-0.5,2)
plt.ylim(-0.5,2)
plt.show()
```

```python
def OR2(x1,x2):
    input = np.array([x1,x2])
    weight = np.array([0.2,0.4])
    bias = -0.1
    value = np.dot(weight,input) + bias
    if value > 0:
        return 1
    else:
        return 0
```

```python
print(OR2(0,0))
print(OR2(1,0))
print(OR2(0,1))
print(OR2(1,1))
```

### OR 눈으로 보기

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.1

y = (-0.2 * x1 - bias) / 0.4
plt.axvline(x=0,color = 'black')
plt.axhline(y=0,color = 'black')

plt.plot(x1,y,'--')
plt.scatter(0,0,color='g',marker='x',s=150)
plt.scatter(1,0,color='red',marker='o',s=150)
plt.scatter(0,1,color='red',marker='o',s=150)
plt.scatter(1,1,color='red',marker='o',s=150)
plt.xlim(-0.5,2)
plt.ylim(-0.5,2)
plt.show()
```

### 2.4.1 XOR게이트

XOR게이트는 배타적 논리합 이라는 논리게이트야. 내가 위에서 예시를 들었던것처럼, (1,0),(0,1)은 '배타적'인 쌍이야 (자기 자신과 같은놈이 있으면 거부해)<br>
이런 서로 다른 입력이 있을떄만 출력이 존재하는 게이트가 XOR게이트야

### XOR게이트의 진리표

### 2.4.2 선형과 비선형

만약 직선으로만 구별하지 않아도 된다면? 에서 XOR문제의 해답은 시작해. 단순하잖아. 그냥 우리가 묶을애들 끼리만 요래조래 잘 구별해주면 되는 문제거든<br>

#### *참고   : 직선의 방정식

선형대수학에서는 $R^n$ 차원의 경우 <b>x</b> = x$_0$ + $t$ $.$ <b>$v$</b> 의 꼴로 표시해. 그렇기 때문에 위에서 퍼셉트론을 직선형태라고 볼 수 있던거야 <br>
$t$는 $n$ 차원의 매개변수, v는 n차원의 방향벡터, x$_0$는 직선 <b>x</b>위의 한 점<br><br>
지금 우리의 경우는 <b>x</b> = y , $v$ = $w_n$,  $t$ = $x_n$, x$_0$ = bias 로 볼 수 있어

value = np.sum(w,x) + bias 꼴로 썻던걸 기억하면 편할꺼야

### 2.5 다층 퍼셉트론

여기서 딥러닝의 기본 개념이 시작돼. 기존 퍼셉트론으로는 XOR문제를 해결할 수 없었는데,<br>
퍼셉트론을 '층을 쌓아' 다층 퍼셉트론으로 만들어준다는 개념으로 XOR 문제를 해결 할 수 있게 된거야.<br>
만드는 방법은 정말 다양하지만, AND,NAND,OR게이트를 통해 XOR게이트를 구현해보자

### 다층 퍼셉트론의 도식화

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```

이게 지금 무슨일이 일어난거냐면,<br>
$[x_1,x_2]$는 입력값이야<br>
$s_1$은 NAND게이트에 $[x_1,x_2]$를 넣었을 때 나온 출력이야. 그래서 [1,1,1,0]이 출력됐어<br>
$s_2$는 OR게이트에 $[x_1,x_2]$를 넣었을 떄 나온 출력이야. 그래서 [0,1,1,1]이 출력됐어<br>
이제 AND게이트에다가 입력값으로 $x_1,x_2$ 대신 $s_1,s_2$를 주면, 우리가 찾던 XOR게이트가 나오게 되는
거야 <br><br>



```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias1 = 0.5
bias2 = 1.5

y1 = ( -1 * x1 + bias1)
y2 = ( -1 * x1 + bias2)
plt.axvline(x=0,color = 'black')
plt.axhline(y=0,color = 'black')

plt.plot(x1,y1,'--',label = 'OR')
plt.plot(x1,y2,'--',label = 'NAND')
plt.scatter(0,0,color='green',marker='x',s=150)
plt.scatter(1,0,color='red',marker='o',s=150)
plt.scatter(0,1,color='red',marker='o',s=150)
plt.scatter(1,1,color='green',marker='x',s=150)
plt.legend()

plt.xlim(-0.5,2)
plt.ylim(-0.5,2)
plt.show()
```

### XOR의 네트워크 구조

XOR을 네트워크로 표현하면 위 그림처럼 나오게 되는데, 한 단계를 거칠때마다 한'층'을 올랐다고 표현할꺼고, 왼쪽부터 차례대로 0층,1층,2층이라부를게.
AND 나 OR 같은 경우, 입력다음에 바로 출력이 있는 단층 퍼셉트론인데, XOR은 2층까지 존재하잖아? 이렇게 층이 여러개인 퍼셉트론을 '다층 퍼셉트론'이라 불러.<br><br>

아니, 분명히 x써있는 층이랑 s써있는 층이랑 y써있는 층이라서 3층같은데 왜 0층부터 시작해놓고 2층이라고 우겨요?<br><br>

음... 그건 처음에 넣어준 x1,x2는 그냥 넣어준다는거고, 가중치를 갖고 실제로 계산하는 과정은 2단계니까 2층 퍼셉트론이라 하기로 하자. 가끔 저것도 3층 퍼셉트론이라 부르는 사람도 있대<br>
근데 파이썬은 숫자를 0부터 세니까, 0부터 시작해야 편리한점이 많아서 0층부터 시작한다고 하자.

```python
def XOR2(x1,x2):
    s1 = NAND(x1,x2)
    s2 = NAND(x1,s1)
    s3 = NAND(s1,x2)
    y = NAND(s2,s3)
    return y
    
```

```python
print(XOR2(0,0))
print(XOR2(1,0))
print(XOR2(0,1))
print(XOR2(1,1))
```

### 2.7 정리

- 퍼셉트론은 입출력을 갖춘 알고리즘. 규칙에 따라 입력에 맞는 값을 출력
- 퍼셉트론은 가중치와 편향을 매개변수로 사용
- 퍼셉트론으로 AND,OR등의 논리게이트를 표현 할 수 있다
- XOR게이트는 단층 퍼셉트론으로 표현 할 수 없다
- 2층 퍼셉트론을 통해 XOR게이트를 표현 할 수 있다
- 단층 퍼셉트론은 직선영역만 표시 가능하고, 다층퍼셉트론은 비선형 영역도 표시 할 수 있다
- 다층퍼셉트론으로 이론상 컴퓨터도 구현 가능하다
